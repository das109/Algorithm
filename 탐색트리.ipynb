```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BST:
    def __init__(self, root):
        self.root = root

    def insert(self, data):
        self.cr_node = self.root
        if self.root.data == None:
            self.root = Node(data)
            return
        while True:
            if data < self.cr_node.data:
                if self.cr_node.left != None:
                    self.cr_node = self.cr_node.left
                else:
                    self.cr_node.left = Node(data)
                    break
            else:
                if self.cr_node.right != None:
                    self.cr_node = self.cr_node.right
                else:
                    self.cr_node.right = Node(data)
                    break

    def search(self, data):
        self.cr_node = self.root
        while self.cr_node:
            if self.cr_node.data == data:
                return data
            elif self.cr_node.data > data:
                self.cr_node = self.cr_node.left
            else:
                self.cr_node = self.cr_node.right
        return False

    def delete(self, data):
        is_search = False
        self.cr_node = self.root
        self.parent = self.root
        while self.cr_node:
            if self.cr_node.data == data:
                is_search = True
                break
            elif data < self.cr_node.data:
                self.parent = self.cr_node
                self.cr_node = self.cr_node.left
            else:
                self.parent = self.cr_node
                self.cr_node = self.cr_node.right
        if is_search == False:
            return False

        tmp = self.cr_node.data
        if self.cr_node.left == None and self.cr_node.right == None:
            if data < self.parent.data:
                self.parent.left = None
            else:
                self.parent.right = None
        
        if self.cr_node.left != None and self.cr_node.right == None:
            if data < self.parent.data:
                self.parent.left = self.cr_node.left
            else:
                self.parent.right = self.cr_node.left
        
        if self.cr_node.left == None and self.cr_node.right != None:
            if data < self.parent.data:
                self.parent.left = self.cr_node.right
            else:
                self.parent.right = self.cr_node.right                

        if self.cr_node.left != None and self.cr_node.right != None:
            self.ch_node = self.cr_node.right
            self.ch_node_parent = self.cr_node.right
            while self.ch_node.left != None:
                self.ch_node_parent = self.ch_node
                self.ch_node = self.ch_node.left
            if self.ch_node.right != None:
                self.ch_node_parent.left = self.ch_node.right
            else:
                self.ch_node_parent.left = None
                
            if data < self.parent.data:
                self.parent.left = self.ch_node
                self.ch_node.right = self.cr_node.right
                self.ch_node.left = self.cr_node.left
            else:
                self.parent.right = self.ch_node
                self.ch_node.left = self.cr_node.left
                self.ch_node.right = self.cr_node.right

        return tmp

    def pre_order(self, root):
        print(root.data, end=' ')
        if (root.left != None) and (root.left.data is not None):
          self.pre_order(root.left)
        if (root.right != None) and (root.right.data is not None):
          self.pre_order(root.right)

    def pr(self):
        print("", end = " ")
        self.pre_order(self.root)


root = Node(None)
li = BST(root)

while (1):
    
    k = list(input().split())

    if (k[0] == 'i'):
      li.insert(int(k[1]))
    elif (k[0] == 'd'):
      result = li.delete(int(k[1]))
      if result == False:
        print("X")
      else:
        print(result)
    elif (k[0] == 's'):
      result = li.search(int(k[1]))
      if result == False:
        print("X")
      else:
        print(result)
    elif (k[0] == 'p'):
      li.pr()
    elif (k[0] == 'q'):
      break# Type or paste your code in this area
  
  # Type or paste your code in this area
```
