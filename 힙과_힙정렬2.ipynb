```python
class Heap:
  def __init__(self, keys):
    self.array = [0]+keys
    self.size = len(self.array)-1

  def rbHeap(self, i):

    if i > self.size:
        return
    
    self.rbHeap(2*i)
    self.rbHeap(2*i+1)
    self.downHeap(self.array[i])


  def downHeap(self, p):

    assert p in self.array
    
    for i, data in enumerate(self.array):
      if data == p:
          p_index = i

    if (len(self.array) <= p_index*2) and (len(self.array) <= (p_index*2+1)):
        return

    b = self.array[p_index*2]

    if (len(self.array) > p_index*2+1):
        if (self.array[p_index*2+1] > b):
          b = self.array[p_index*2+1]

    if p > b:
        return

    for i, data in enumerate(self.array):
        if data == b:
          b_index = i
      

    self.array[p_index] = self.array[b_index]
    self.array[b_index] = p
    self.downHeap(self.array[b_index])


  def bHeap(self):

    for i in range(self.size//2, 0, -1):
      self.downHeap(self.array[i])
    return self.pr()



  def pr(self):
    print("", end = " ")
    for data in self.array[1:]:
      print(data, end = " ")
    print()
    




n = int(input())

k = [int(i) for i in input().split()]
assert len(k) == n

Heap = Heap(k)
Heap.rbHeap(1)
Heap.bHeap()
```
