```python
class Heap:
  def __init__(self):
    self.array = [0, None]


  def upHeap(self, p):

    if p == self.array[1]:
      return
    
    for i, data in enumerate(self.array):
      if data == p:
          p_index = i
    if p < self.array[p_index//2]:
        return

    self.array[p_index] = self.array[p_index//2]
    self.array[p_index//2] = p
    self.upHeap(self.array[p_index//2])


  def downHeap(self, p):

    assert p in self.array
    
    for i, data in enumerate(self.array):
      if data == p:
          p_index = i

    if (len(self.array) <= p_index*2) & (len(self.array) <= (p_index*2+1)):
        return

    b = self.array[p_index*2]

    if (len(self.array) > p_index*2+1):
        if (self.array[p_index*2+1] > b):
          b = self.array[p_index*2+1]

    if p > b:
        return

    for i, data in enumerate(self.array):
        if data == b:
          b_index = i
      

    self.array[p_index] = self.array[b_index]
    self.array[b_index] = p
    self.downHeap(self.array[b_index])


  def insert(self, data):

    assert len(self.array) <100

    if self.array[1] == None:
        self.array[1] = data
    
    else:
        self.array.append(data)

    p = self.array[-1]
    self.upHeap(p)

    return 0

  def delete(self):

    root = self.array[1]
    temp = self.array[-1]
    self.array.remove(root)
    self.array.remove(0)
    self.array.remove(temp)

    self.array = [0, temp] + self.array
    self.downHeap(self.array[1])

    return root


  def pr(self):
    print("", end = " ")
    for data in self.array[1:]:
      print(data, end = " ")
    print()
    



Heap = Heap()

while (1):

  k = list(input().split())

  if (k[0] == 'i'):
    print(Heap.insert(int(k[1])))
  elif (k[0] == 'd'):
    print(Heap.delete())
  elif (k[0] == 'p'):
    Heap.pr()
  elif (k[0] == 'q'):
    break



```
